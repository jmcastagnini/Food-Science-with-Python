```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from anova_one_way import anova1
from graphic_letters import graphic_letters
```

```python
df_result=anova1(df='color2.csv', treatment=1,col=3)  # Statistical analysis 
df_result
```

```python
df = pd.read_csv('color2.csv', sep=';')  # Read the data frame
#df.replace(to_replace='P-05', value='P-5', inplace=True)  # Replace some values
```

```python
fig, ax = plt.subplots(1,3, figsize=(14,5))  # Create a figure of 1 row and 3 columns
sns.set_theme(style="whitegrid")

# First place
ax[0].set_title('Luminosità')  # Title
chart=sns.barplot(ax=ax[0],x='Sample', y='L', data=df, ci='sd');  # Make the graphic
graphic_letters(chart=chart,df=df_result,analysis='L',xytext=(3,6),fontsize=13)  # Include the statistical letters

# Second place
ax[1].set_title('Croma')
#ax[1].set_ylim(0, -3.5)  # Change the limits of y axis
chart=sns.barplot(ax=ax[1],x='Sample', y='C', data=df, ci='sd');
graphic_letters(chart=chart,df=df_result,analysis='a',xytext=(3,6),fontsize=13)

# Third place
ax[2].set_title('Hue')
chart=sns.barplot(ax=ax[2],x='Sample', y='h', data=df, ci='sd');
graphic_letters(chart=chart,df=df_result,analysis='b',xytext=(3,6),fontsize=13)
```
